```python
f = open("input.txt", "r")

inp = f.readlines()

inp = [rel[:-1] for rel in inp]
```

```python
inp = [rel.split(')') for rel in inp]
```

```python
[*rel for rel in inp]
```

```python
from functools import reduce
import numpy

uniqs = set(reduce(numpy.append, inp))
```

```python
# ['497' == rel for _, rel in inp]
inp[[i for i, x in enumerate(['497' == rel for _, rel in inp]) if x][0]][0]
```

```python
sums = 0
for planet in uniqs:
    try:
        next_p = inp[[i for i, x in enumerate([planet == rel for _, rel in inp]) if x][0]][0]
    except:
        continue
    sums+=1
    while 1:
        try:
            next_p = inp[[i for i, x in enumerate([next_p == rel for _, rel in inp]) if x][0]][0]
            sums+=1
        except:
            break
            
```

```python
path_u = []
planet = 'YOU'
try:
    next_p = inp[[i for i, x in enumerate([planet == rel for _, rel in inp]) if x][0]][0]
    path_u.append(next_p)
except:
    pass
sums+=1
while 1:
    try:
        next_p = inp[[i for i, x in enumerate([next_p == rel for _, rel in inp]) if x][0]][0]
        path_u.append(next_p)
    except:
        break
```

```python
path_u.index('RQZ')
```

```python
path_san = []
planet = 'SAN'
try:
    next_p = inp[[i for i, x in enumerate([planet == rel for _, rel in inp]) if x][0]][0]
    path_san.append(next_p)
except:
    pass
sums+=1
while 1:
    try:
        next_p = inp[[i for i, x in enumerate([next_p == rel for _, rel in inp]) if x][0]][0]
        path_san.append(next_p)
    except:
        break
```

```python
path_san[0]
```

```python
cost_m = 300000
for pl in path_u:
    if pl in path_san:
        cost = path_u.index(pl) + path_san.index(pl)
        if cost<cost_m:
            cost_m = cost
```
